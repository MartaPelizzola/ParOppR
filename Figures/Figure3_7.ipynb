## Plotting the results

```r
library(dplyr)
library(ggplot2)
```

```r
opp = unlist(read.table(file = "pcawg_pilot_c7_partition0_num_sites.tsv", sep = '\t', header = F))
# load(paste0("../../results/rdata_overlap/brcacounts7_start", startID, ".RData")) 
```

```r
# plot function 

plotres = function(dat_full,dat_full_opp, dat_di, dat_di_opp, dat_mono, dat_mono_opp){
    data_all = rbind(dat_full,dat_full_opp, dat_di, dat_di_opp, dat_mono, dat_mono_opp)

    data_all$Opportunity = factor(data_all$opp, levels = c("Yes", "No"))
    data_all$param = factor(data_all$param, levels = c("Mono","Di", "Standard"), labels = c("Additive", "Interaction", "Standard"))

    data_all3 = data_all[data_all$sigNo < 17,]
    data_all2 = data_all3 %>% group_by(param,Opportunity) %>% mutate(min_val = min(error), min_no = sigNo[which.min(error)])

    g1 = ggplot(data_all2, aes(x = sigNo, y = error, color = param, shape = Opportunity, linetype = Opportunity))+
    geom_point()+
    geom_line()+
    scale_color_manual(name = 'Parametrization', values = c('#AC0136',"#FF9912","#27408B"))+
    theme_bw()+
    geom_point(aes(x = min_no, y = min_val), size = 4, alpha = 0.1)+
    xlab('Number of signatures')

    return(g1)
}
```
